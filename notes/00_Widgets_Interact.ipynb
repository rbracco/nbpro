```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display, Audio
import numpy as np
from functools import partial
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x
```

## ipywidgets interact

```python
interact(f, x=10);
```

```python
def make_tone(freq, clip_length=1, sr=16000):
    t = np.linspace(0, clip_length, int(clip_length*sr), endpoint=False)
    print(np.expand_dims(t, 0).shape)
    sig = 0.1*np.sin(2*np.pi*freq*t)
    display(plt.plot(t[0:1000], sig[0:1000]))
    display(Audio(sig, rate=sr))
    return 0.1*np.sin(2*np.pi*freq*t)
```

```python
interact_manual(make_tone, freq=100, continuous_update=False);
```

```python
interact(make_tone, freq=100, clip_length=fixed(1), sr=fixed(16000));
```

```python
interact(make_tone, 
         freq=widgets.IntSlider(min=0, max=100, step=10, value=20),
        clip_length=fixed(1), 
         sr=fixed(16000));
```

## Argument types

```python
interact(make_tone, 
         freq=widgets.IntSlider(min=0, max=100, step=10, value=20),
        clip_length=fixed(1), 
         sr=fixed(16000));
```

```python
interact(make_tone, freq=(0,100,10), clip_length=fixed(1), sr=fixed(16000));
```

```python
# note this is interactive, not interact. interactive lets you access the return value
# of the function being hooked into using .result
w = interactive(make_tone, freq=(0,100,10), clip_length=fixed(1), sr=fixed(16000))
```
